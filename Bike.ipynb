```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
num_vars = ["temp","atemp","hum","windspeed","season", "yr","mnth","holiday","weekday","workingday","weathersit","casual", "registered","cnt"]
df_train[num_vars] = scaler.fit_transform(df_train[num_vars])
```
